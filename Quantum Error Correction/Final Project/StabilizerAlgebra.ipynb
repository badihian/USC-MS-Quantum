```python
def valid_stab(stab):
  for i in stab:
    if i not in {'X','Y','Z','I'}:
      return False
  return True
################################################################################
def normlength(stab1, stab2):
  if valid_stab(stab1) == False:
    return(False)

  if valid_stab(stab2) == False:
    return(False)

  if (len(stab1) > len(stab2)): #length matching with I
    diff = len(stab1) - len(stab2)
    stab2 = stab2 + ('I'*diff)

  if (len(stab1) < len(stab2)): #length matching with I
    diff = len(stab2) - len(stab1)
    stab1 = stab1 + ('I'*diff)
  return([stab1, stab2])
################################################################################
def commute(stab1,stab2):
  stab1 = normlength(stab1, stab2)[0]
  stab2 = normlength(stab1, stab2)[1]

  count = 0
  for p in range(len(stab1)+1):
    if ((stab1[p-1:p] != 'I' and stab2[p-1:p] != 'I') and stab1[p-1:p] != stab2[p-1:p]): #no identities, not matching paulis
      count += 1
      # print(p) #anti-commuting qubits
  if (count%2 == 0):
    return(True)
  return(False)
################################################################################
def pauli_mult(p1,p2):
  if p1 == 'I':
    return([1,p2])
  if p2 == 'I':
    return([1,p1])
  if p1==p2:
    return([1,'I'])
  if (p1 == 'X' and p2 == 'Y'):
    return([1j,'Z'])
  if (p1 == 'Y' and p2 == 'Z'):
    return([1j,'X'])
  if (p1 == 'Z' and p2 == 'X'):
    return([1j,'Y'])
  if (p1 == 'Y' and p2 == 'X'):
    return([-1j,'Z'])
  if (p1 == 'Z' and p2 == 'Y'):
    return([-1j,'X'])
  if (p1 == 'X' and p2 == 'Z'):
    return([-1j,'Y'])
  else:
    return("invalid")
################################################################################  
def stab_mult(stab1, stab2):
  stab1 = normlength(stab1, stab2)[0]
  stab2 = normlength(stab1, stab2)[1]

  new_stab = ''
  phase = 1
  for p in range(1, len(stab1)+1):
    new_stab += pauli_mult(stab1[p-1:p],stab2[p-1:p])[1] #multiply paulis, add to pauli string
    phase *= pauli_mult(stab1[p-1:p],stab2[p-1:p])[0] #multiply phase to current phase
  return([phase, new_stab]) 
################################################################################
def hadamard(stab, qubit):
  if stab[qubit-1] == 'X':
    newq = 'Z'
  if stab[qubit-1] == 'Z':
    newq = 'X'
  else:
    newq = stab[qubit-1]
  return(stab[:qubit-1] + newq + stab[qubit:])
################################################################################
def cnot(stab1, stab2, qubit): #cnot from first to second stabilizer
  stab1 = normlength(stab1, stab2)[0]
  stab2 = normlength(stab1, stab2)[1]

  p1 = stab1[qubit-1]
  p2 = stab2[qubit-1]
  p2new = p2
  phase = 1
  if p1 == 'X' or p1 == 'Y':
    phase *= pauli_mult('X',p2)[0]
    p2new = pauli_mult('X',p2)[1]

  if p2 == 'Z' or p2 == 'Y':
    phase *= pauli_mult(p1,'Z')[0]
    p1 = pauli_mult(p1,'Z')[1]
  p2 = p2new
  return([phase,stab1[:qubit-1] + p1 + stab1[qubit:],stab2[:qubit-1] + p2 + stab2[qubit:]])
################################################################################
  




#print(commute('IXXZY','IZXI'))
#print(pauli_mult('X','Y')[1]+pauli_mult('Y','Y')[1])
#print(stab_mult('IXXZY','IZXI'))
#print(hadamard('IXIZYIXYZ',4))
print(cnot('IXZZX','YZYX',2))
print(cnot('IXZZX','YZYX',3))
print(cnot('IXZZX','YZYX',5))
print(cnot('IXZZX','YZYX',1))
print(cnot('YZYX','IZZXZ',3))
print(cnot('IXZZX','YZYXII',4))
print(commute(cnot('IXZZX','YZYXII',4)[1],cnot('IXZZX','YZYXII',4)[2]))
```
