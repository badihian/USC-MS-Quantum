```python
import operator as op
from functools import reduce

def comb(n, r):
    r = min(r, n-r)
    numer = reduce(op.mul, range(n, n-r, -1), 1)
    denom = reduce(op.mul, range(1, r+1), 1)
    return numer // denom

def hamming(n,k,t):
  count = 0
  valid = False
  for j in range(t+1):
    count += (comb(n,j)*(3**j))
  count -= 2**(n-k)
  if count > 0:
    valid = True
  #print(count)
  return(valid)

def nkleast(max_n):
  for n in range(5, max_n+1):
    for k in range(0,n):
      t=1
      while(hamming(n,k,t) == False):
        t+=1
      print("[[",n,",",k,",",t,"]]", end='  ')
    print('\n\n')
  return("These are the [[n,k,t]] for the minimum weight (t) of each n/k combo. Note that minimum distance d=2t+1")

print(nkleast(25))


```
